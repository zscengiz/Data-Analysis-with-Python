```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

airline_bumping = pd.read_csv("airline_bumping.csv")

file_path = 'avoplotto.pkl'
avocados = pd.read_pickle(file_path)
print(avocados.head())
```

**Visualizing You Data**

```python
# Print the head of the avocados dataset
print(avocados.head())

# For each avocado size group, calculate the total number sold, storing as nb_sold_by_size
nb_sold_by_size = avocados.groupby('size')['nb_sold'].sum()

# Create a bar plot of the number of avocados sold by size
nb_sold_by_size.plot(kind='bar', color='skyblue')
plt.title('Number of Avocados Sold by Size')
plt.xlabel('Avocado Size')
plt.ylabel('Total Number of Avocados Sold')

# Show the plot
plt.show()
```

```python
# Get the total number of avocados sold on each date
nb_sold_by_date = avocados.groupby('date')['nb_sold'].sum()

# Create a line plot of the number of avocados sold by date
nb_sold_by_date.plot(kind='line', marker='o', linestyle='-', color='skyblue')
plt.title('Number of Avocados Sold by Date')
plt.xlabel('Date')
plt.ylabel('Total Number of Avocados Sold')

# Show the plot
plt.show()
```

```python
# Scatter plot of avg_price vs. nb_sold with title
avocados.plot(kind='scatter', x='avg_price', y='nb_sold', title='Scatter Plot of Avg Price vs Number of Avocados Sold')

# Show the plot
plt.show()
```

```python
# Scatter plot with nb_sold on the x-axis and avg_price on the y-axis
avocados.plot(kind='scatter', x='nb_sold', y='avg_price', title='Number of avocados sold vs. average price')

# Show the plot
plt.show()
```

```python
# Histogram of conventional avg_price 
avocados[avocados["type"] == "conventional"]["avg_price"].hist()

# Histogram of organic avg_price
avocados[avocados["type"] == "organic"]["avg_price"].hist()

# Add a legend
plt.legend(["conventional", "organic"])

# Show the plot
plt.show()
```

```python
# Modify histogram transparency to 0.5 for conventional avocados
avocados[avocados["type"] == "conventional"]["avg_price"].hist(alpha=0.5)

# Modify histogram transparency to 0.5 for organic avocados
avocados[avocados["type"] == "organic"]["avg_price"].hist(alpha=0.5)

# Add a legend
plt.legend(["conventional", "organic"])

# Show the plot
plt.show()
```

```python
# Modify bins to 20 for conventional avocados
avocados[avocados["type"] == "conventional"]["avg_price"].hist(alpha=0.5, bins=20)

# Modify bins to 20 for organic avocados
avocados[avocados["type"] == "organic"]["avg_price"].hist(alpha=0.5, bins=20)

# Add a legend
plt.legend(["conventional", "organic"])

# Show the plot
plt.show()
```

**Creating DataFrames**
